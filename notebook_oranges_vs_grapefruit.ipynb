```python
import pandas as pd
df = pd.read_csv('citrus.csv')
```

```python
df.info()
```

```python
df.head(10000)
```

```python
# ANN cannot process strings, so the value in the label (name) column must be converted to numeric value
df.name[df.name == 'orange'] = 0
df.name[df.name == 'grapefruit'] = 1
```

```python
# the model still can not process the dataset because it is still in the form of a dataframe
# the dataset must be in the form of an array so that it can be processed by the model
# attribute 'values' returns the converted numpy array from the dataframe

dataset = df.values
```

```python
dataset
```

```python
# separate the attributes and the labels
# from data exploration, we know that column 1 is a label, columns 2-5 are attributes

# select the last 5 columns as attributes
X = dataset[:,1:6]
    # the number before the comma is used to select rows in the dataframe
    # the number after the comma is used to select columns in the dataframe

# create a numpy array containing only labels
y = dataset[:,0]
```

```python
# normalization
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
# split the dataset into train and test sets
# the ratio is 70% : 30%

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
# earlier, we changed the values in the column 'name' to numeric numbers,
# which means that we changed the labels to the boolean data type,
# then we need to change the data type to float32

import numpy as np
Y_train = Y_train.astype(np.float32)
Y_test = Y_test.astype(np.float32)

# the data processing step has been completed
```

```python
# build an ANN model architecture

from keras.models import Sequential
from keras.layers import Dense

model = Sequential([Dense(32, activation='relu', input_shape=(5,)),
                    Dense(32, activation='relu'),
                    Dense(1, activation='sigmoid'),])  # sigmoid is used to map probabilities from 0 to 1
```

```python
# determine the optimizer and loss function of the model

model.compile(optimizer='sgd',            # stochastic gradient descent ('sgd') is very common and suitable for small datasets
              loss='binary_crossentropy', # 'binary_crossentropy' is used due to 2 class problem (binary)
              metrics=['accuracy'])       # add parameter metrics=['accuracy'] to display accuracy in each model training process
```

```python
# train the model
model.fit(X_train, Y_train, epochs=100)
```

```python
# evaluate the model
model.evaluate(X_test, Y_test)  # (loss, accuracy)
```
